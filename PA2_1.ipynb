```python
# Import libraries
import pandas as pd
import altair as alt
import csv
```

```python
# Data retrieved from https://www.kaggle.com/crawford/80-cereals?select=cereal.csv
# Read the data and put it in a panda's dataframe

attributes = []
records = []
with open('cereal.csv') as inFile:
    reader = csv.reader(inFile)
    for row in reader:
        if (row[0] == "name"):
            attributes = row
            continue
        records.append(row)

df = pd.DataFrame(records, columns = attributes)
df
```

```python

PAD = 20
```

```python
scatterPlot =alt.Chart(df).mark_point(size=20).encode(
    x=alt.X('carbo:Q', 
        sort=alt.EncodingSortField('carbo', order='descending'), 
        axis=alt.Axis(title='Carbohydrates (g)')
    ),
    y=alt.Y('calories:Q',
        sort=alt.EncodingSortField('calories', order='descending'), 
        axis=alt.Axis(title='Calories')
    )
).properties(
    title='The Affect of Carbohydrates on Calories',
    width=300,
    height=220
)

scatterPlot + scatterPlot.transform_regression('carbo', 'calories', 
    extent=[-10,25] ,
    method="linear"
).mark_line(color="red")
```

```python
alt.Chart(df).mark_circle().encode(
    x = alt.X('calories:Q', 
        axis=alt.Axis(title='Calories'),
        bin=True),
    y = alt.Y('rating:Q', 
        axis=alt.Axis(title='Rating'),
        bin=True),
    size='count()'
).properties(
    title='The Affect of Calories on Ratings (Binned)',
    width=300,
    height=220
).configure(
    padding={"left": PAD, "top": PAD, "right": PAD, "bottom": PAD}
)
```

```python
alt.Chart(df).mark_boxplot().encode(
    x = alt.X('mfr:O', 
        axis=alt.Axis(title='Manufacturer')),
    y = alt.Y('rating:Q', 
        axis=alt.Axis(title='Rating'))
).properties(
    title='A Comparison of Each Manufacturer\'s Ratings',
    width=300,
    height=220
).configure(
    padding={"left": PAD, "top": PAD, "right": PAD+20, "bottom": PAD}
)
```
